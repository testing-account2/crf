```python
import sys
sys.path

path="c:/users/ssong/appdata/local/programs/python/python37-32/lib/site-packages"
path2="s:/acrf.pdf"
sys.path.append(path)

import PyPDF2, traceback
input1 = PyPDF2.PdfFileReader(open(path2, "rb"))
npages = input1.getNumPages()

d={} 
for i in range(npages) :
        page0 = input1.getPage(i) 
        try :
            for x in page0['/Annots']:

                #print ('varname: %s pages on:%s'%([x.getObject()][0]['/Contents'],i))
                #print ('')
                if  [x.getObject()][0]['/Contents']  in d.keys() and i not in d[[x.getObject()][0]['/Contents']]:
                        #print (d[[x.getObject()][0]['/Contents']])
                        #v=d[[x.getObject()][0]['/Contents']];v.append(i)
                        d[[x.getObject()][0]['/Contents']].append(i)
                                           
                else:
                    #d.update([([x.getObject()][0]['/Contents'],i)] )
                    d[[x.getObject()][0]['/Contents']]=[i]
             
        except : 
        # there are no annotations on this page
            pass
    
```

```python
import pandas as pd
d2=pd.DataFrame([d]).T
d2.columns=['pageno']
d2.to_csv("c:/file.csv")
d2
```
